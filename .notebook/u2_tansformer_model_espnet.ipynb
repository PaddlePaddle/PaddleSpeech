```python
# dump torch model to paddle
#state_dict = model.state_dict()
paddle_state_dict = {}

for n, p in state_dict.items():
    if 'encoder' not in n:
        continue 
    print(f'-> {n}')
    
    
    name_change=True
    if 'norm.running_mean' in n:
        new_n = n.replace('norm.running_', 'norm._')
    elif 'norm.running_var' in n:
        new_n = n.replace('norm.running_var', 'norm._variance')
    else:
        name_change=False
        new_n = n
    if name_change:
        print(f"{n} -> {new_n}")
        
        
    p = p.cpu().detach().numpy()
    if n.endswith('weight') and p.ndim == 2:
        new_p = p.T
        print(f"{n}: {p.shape} -> {new_p.shape}")
    else:
        new_p = p
        
    if 'global_cmvn.mean' in n:
        print(p, p.dtype)
        
    paddle_state_dict[new_n] = new_p
    
# np.savez('/workspace/DeepSpeech-2.x/.notebook/model',
#        state=paddle_state_dict)
```

```python
model.set_state_dict(paddle_state_dict)
```

```python
e_state = model.encoder.state_dict()
for key, value in e_state.items():
    if 'concat_linear' in key:
        continue
    if not np.allclose(value.numpy(), paddle_state_dict['encoder.' + key]):
        print(key)
```
